```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns',None)
```

```python
credit = pd.read_csv('credits.csv')
credit
```

```python
meta=pd.read_csv('movies_metadata.csv')
meta.head(5)
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'],errors='coerce')
```

```python
meta['year'] = meta['release_date'].dt.year
```

```python
meta
```

```python
meta['year'].value_counts().sort_index()
```

```python
new_meta = meta.loc[meta.year == 2017,['genres','id','title','year']]
new_meta
```

```python
new_meta['id']=new_meta['id'].astype('int')
```

```python
data = pd.merge(new_meta,credit,on='id')
data
```

```python
import ast

data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data['genres']
```

```python
def make_genrelist(x):
    gen = []
    st = ' '
    for i in x:
        if i.get('name') == 'Science Fiction':
            gen.append('Sci-Fi')
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].map(lambda x:make_genrelist(x))
data
```

```python
data['crew'].loc[100]
```

```python
def get_actor1(x):
    cast=[]
    for i in x:
        cast.append(i.get('name'))
    if cast == [] or len(cast)<=1:
        return np.NAN
    else:
        return cast[0]
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
data
```

```python
def actor_2(x):
    cast=[]
    for i in x:
        cast.append(i.get('name'))
    if cast == [] or len(cast)<=1:
        return np.NAN
    else:
        return cast[1]
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: actor_2(x))
data
```

```python
def actor_3(x):
    cast=[]
    for i in x:
        cast.append(i.get('name'))
    if cast == [] or len(cast)<=1:
        return np.NAN
    else:
        return cast[1]
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: actor_3(x))
data
```

```python
def  get_director(x):
    dt = []
    st = ' '
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return st.join(dt)
```

```python
data['director_name'] = data['crew'].map(lambda x: get_director(x))
data
```

```python
movie  =  data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','title','genres_list']]
movie
```

```python
movie.isna().sum()
```

```python
movie.dropna(how='any',inplace=True)
movie
```

```python
movie=movie.rename(columns={'genres_list':"genres"})
movie=movie.rename(columns={'title':"movie_title"})
movie['movie_title']=movie['movie_title'].str.lower()
movie
```

```python
movie['comb'] = movie['actor_1_name'] + " "+ movie['actor_2_name'] + " "+ movie['actor_3_name'] + " "+ movie['director_name'] + " " +movie['genres']+ " " + movie['movie_title']+ " " + movie['director_name']
movie
```
